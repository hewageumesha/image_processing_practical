**Importing required libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
```

**Reading and Displaying image**

```python
#Read the image from file
image = cv2.imread('image.jpg')

#Display the image in the window
cv2_imshow(image)
```

**Converting between color spaces**

```python
#convrt the grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Save the grayscale image
cv2.imwrite('gray_image.jpg', gray_image)

#show grayscale image
cv2_imshow(gray_image)
```

**Resize the image**

```python
resized_image = cv2.resize(image, (300,200))
cv2.imwrite('resized_image.jpg', resized_image)
cv2_imshow(resized_image)
```

**Rotate image**

```python
#get the image's height and width
(h, w) = image.shape[:2]

#define the center of the image(using // round the result down to the nearest whole number)
center = (w // 2, h // 2)

#creates a rotation matrix that rotates the image 90 degrees counterclockwise around its center.
#1.0 -> keep original size
matrix = cv2.getRotationMatrix2D(center , 180, 1.0)

#perform the rotation
rotate_image = cv2.warpAffine(image, matrix, (w,h))

#save the rotated image
cv2.imwrite('rotated_image.jpg', rotate_image)

#show image
cv2_imshow(rotate_image)
```

**Gaussian Blur**

```python
#apply gaussian blur
blurred_image = cv2.GaussianBlur(image, (15,15), 0)

#save the blurred image
cv2.imwrite('blurred_image.jpg', blurred_image)

#show image
cv2_imshow(blurred_image)
```
